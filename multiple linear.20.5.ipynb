```python
from sklearn.linear_model import LinearRegression
# Sample input features (X) and target variable (y)
X=[[1,2,3],[4,5,6],[7,8,9]]
y=[10,20,30]
# create a linear regression model 
model=LinearRegression()
# Train the model 
model.fit(X,y)
# predict using the trained model 
new_data=[[2,3,4],[5,6,7]]
predictions=model.predict(new_data)
# prin the predictins
for i, prediction in enumerate(predictions):
    print(f"prediction{i+1}:{prediction}")
```

```python
#polynomial linear regression 
import numpy as np
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
# Sample imput data 

X=np.array([1,2,3,4,5]).reshape(-1,1)
# independent Variable 
print(X)
y=np.array([3,6,8,10,15])

print(y)
# Dependent variable 
# Transforn input data to include polynomial features 
poly_features=PolynomialFeatures(degree=2)
```

```python
X_poly=poly_features.fit_transform(X)
print(X_poly)
```

```python
# Create a linear regression model 
model=LinearRegression()
# Train the model 
model.fit(X_poly, y)
```

```python
# Generate prediction using the trained model 
X_pred=np.linspace(1,5,100).reshape(-1,1)
X_pred_poly=poly_features.transform(X_pred)
y_pred=model.predict(X_pred_poly)
```

```python
#plot the original data and the fitted polynomial curve
plt.scatter(X,y,color='b',label='Actual Data')
plt.plot(X_pred,y_pred,color='r',label='Polynomial Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Linear Regression')
plt.legend() 
plt.show()
```
